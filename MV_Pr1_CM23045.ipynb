```python
from google.colab import files

uploaded = files.upload()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_path = list(uploaded.keys())[0]

img = cv2.imread(image_path)

img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(6,6))
plt.imshow(img)
plt.title("Original CCTV Image")
plt.axis("off")
plt.show()
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

equalized = cv2.equalizeHist(gray)

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
plt.imshow(gray, cmap='gray')
plt.title("Original Grayscale")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(equalized, cmap='gray')
plt.title("Histogram Equalized")
plt.axis("off")

plt.show()
```

```python
bright = cv2.convertScaleAbs(img, alpha=1.3, beta=40)

plt.figure(figsize=(6,6))
plt.imshow(bright)
plt.title("Brightness Enhanced")
plt.axis("off")
plt.show()
```

```python
gaussian = cv2.GaussianBlur(bright,(5,5),0)

plt.figure(figsize=(6,6))
plt.imshow(gaussian)
plt.title("Gaussian Filter")
plt.axis("off")
plt.show()
```

```python
median = cv2.medianBlur(bright,5)

plt.figure(figsize=(6,6))
plt.imshow(median)
plt.title("Median Filter")
plt.axis("off")
plt.show()
```

```python
bilateral = cv2.bilateralFilter(bright,9,75,75)

plt.figure(figsize=(6,6))
plt.imshow(bilateral)
plt.title("Bilateral Filter")
plt.axis("off")
plt.show()
```

```python
plt.figure(figsize=(16,8))

titles = [
    "Original",
    "Histogram Equalization",
    "Brightness Enhanced",
    "Gaussian Filter",
    "Median Filter",
    "Bilateral Filter"
]

images = [
    img,
    cv2.cvtColor(equalized, cv2.COLOR_GRAY2RGB),
    bright,
    gaussian,
    median,
    bilateral
]

for i in range(6):
    plt.subplot(2,3,i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis("off")

plt.tight_layout()
plt.show()
```

```python
final = cv2.cvtColor(bilateral, cv2.COLOR_RGB2BGR)

cv2.imwrite("Enhanced_CCTV_Image.jpg", final)

print("Enhanced image saved successfully.")
```
